## Data Cleaning and Transformation

```python
from pyspark.sql.functions import from_unixtime, col, date_trunc, max, min, sum, first, last

# Convert the Timestamp column to a proper Spark timestamp type
df = df.withColumn("Timestamp", from_unixtime(col("Timestamp")).cast("timestamp"))

print("Schema with updated Timestamp type:")
df.printSchema()
```

```python
# Create a daily DataFrame for EDA
df_daily = df.withColumn("Date", date_trunc("day", col("Timestamp")))

df_daily = df_daily.groupBy("Date").agg(
    first("Open").alias("Open"),
    max("High").alias("High"),
    min("Low").alias("Low"),
    last("Close").alias("Close"),
    sum("Volume").alias("Volume")
).orderBy("Date")

# Drop any days that might have no data (optional, but good practice)
df_daily = df_daily.na.drop()

print("\nDaily DataFrame schema:")
df_daily.printSchema()

print("\nFirst 10 rows of Daily DataFrame:")
df_daily.show(10)

print("\nSummary statistics of Daily DataFrame:")
df_daily.describe().show()

print("\nNull value counts per column in Daily DataFrame:")
null_counts_daily = df_daily.select([spark_sum(col(c).isNull().cast("integer")).alias(c) for c in df_daily.columns])
null_counts_daily.show()
```

```python
import networkx as nx
import numpy as np

# Visualization 4: Feature Correlation Network

# Calculate the correlation matrix (excluding 'Date' and NaN rows from rolling calculation)
correlation_matrix = df_daily_pd.drop(columns=['Date', 'Daily_Return', 'Volatility_30d']).corr()

# Create a graph from the correlation matrix
# Use a threshold to only show strong correlations (e.g., absolute correlation > 0.8)
threshold = 0.8
G = nx.Graph()

for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        feature1 = correlation_matrix.columns[i]
        feature2 = correlation_matrix.columns[j]
        correlation_value = correlation_matrix.iloc[i, j]
        if abs(correlation_value) > threshold:
            G.add_edge(feature1, feature2, weight=correlation_value)

# Draw the network
pos = nx.spring_layout(G) # positions for all nodes

plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='skyblue', font_size=10, width=[abs(d['weight'])*3 for (u, v, d) in G.edges(data=True)], edge_color=[d['weight'] for (u, v, d) in G.edges(data=True)], edge_cmap=plt.cm.coolwarm, connectionstyle='arc3,rad=0.1')
labels = nx.get_edge_attributes(G, 'weight')
edge_labels = {k: f"{v:.2f}" for k, v in labels.items()}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

plt.title('Feature Correlation Network (Absolute Correlation > 0.8)')
plt.show()
```

**Insight: Feature Correlation Network**

The feature correlation network visualization reveals the strength and direction of linear relationships between the numerical features in the daily dataset. As expected in financial time series data, the Open, High, Low, and Close prices are all very strongly positively correlated with each other, forming a tightly connected cluster in the network. This indicates that these price points tend to move together. The Volume feature, while connected to the price features, generally shows weaker correlations compared to the inter-price correlations, suggesting that while volume is related to price movements, the relationship is not as consistently linear as the relationships between the different price measures themselves.

## Feature Engineering

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import lag, col, avg, stddev, when

# Define a window partitioned by nothing and ordered by Timestamp for time-series operations
windowSpec = Window.orderBy("Timestamp")

# Create Target Variables
# Regression Target: Next minute's closing price
df = df.withColumn("Close_t+1", lag(col("Close"), -1).over(windowSpec))

# Classification Target: Direction of price change (1 for up, 0 for down or no change)
df = df.withColumn("Direction_t+1", when(col("Close_t+1") > col("Close"), 1).otherwise(0))

# Create Feature Columns
# Lag Features
df = df.withColumn("Close_Lag1", lag(col("Close"), 1).over(windowSpec))
df = df.withColumn("Close_Lag5", lag(col("Close"), 5).over(windowSpec))
df = df.withColumn("Close_Lag15", lag(col("Close"), 15).over(windowSpec))
df = df.withColumn("Volume_Lag1", lag(col("Volume"), 1).over(windowSpec))
df = df.withColumn("Volume_Lag5", lag(col("Volume"), 5).over(windowSpec))
df = df.withColumn("Volume_Lag15", lag(col("Volume"), 15).over(windowSpec))

# Rolling Window Features
df = df.withColumn("Close_MA30", avg(col("Close")).over(windowSpec.rowsBetween(-30, -1)))
df = df.withColumn("Close_MA60", avg(col("Close")).over(windowSpec.rowsBetween(-60, -1)))
df = df.withColumn("Volume_StdDev30", stddev(col("Volume")).over(windowSpec.rowsBetween(-30, -1)))

# Final Cleanup: Drop rows with nulls introduced by window functions
df_engineered = df.na.drop()

print("Schema of DataFrame with engineered features:")
df_engineered.printSchema()

print("\nFirst 10 rows of DataFrame with engineered features:")
df_engineered.show(10)

print("\nSummary statistics of DataFrame with engineered features:")
df_engineered.describe().show()

print("\nNull value counts per column in DataFrame with engineered features:")
null_counts_engineered = df_engineered.select([spark_sum(col(c).isNull().cast("integer")).alias(c) for c in df_engineered.columns])
null_counts_engineered.show()

# Select final columns for modeling
feature_columns = ["Timestamp", "Open", "High", "Low", "Close", "Volume",
                   "Close_Lag1", "Close_Lag5", "Close_Lag15",
                   "Volume_Lag1", "Volume_Lag5", "Volume_Lag15",
                   "Close_MA30", "Close_MA60", "Volume_StdDev30"]

target_columns = ["Close_t+1", "Direction_t+1"]

df_final = df_engineered.select(feature_columns + target_columns)

print("\nSchema of final DataFrame for modeling:")
df_final.printSchema()

print("\nFirst 10 rows of final DataFrame for modeling:")
df_final.show(10)
```

## Insights from Feature Engineering

The feature engineering step was crucial for transforming the raw 1-minute Bitcoin time-series data into a format suitable for predictive modeling. The creation of target variables (`Close_t+1` and `Direction_t+1`) directly addresses the problem of predicting future price movements, framing it as both a regression and a classification task.

The engineered features provide the models with historical context and indicators of market momentum and volatility:

- **Lag Features (Close_Lag1, Close_Lag5, Close_Lag15, Volume_Lag1, Volume_Lag5, Volume_Lag15)**: These capture the recent past values of price and volume. In time-series analysis, past values are often strong predictors of future values, reflecting short-term trends and persistence.
- **Rolling Window Features (Close_MA30, Close_MA60, Volume_StdDev30)**: These provide smoothed versions of the data and measures of variability over longer periods. Moving averages help identify underlying trends by filtering out short-term noise, while rolling standard deviation of volume gives an indication of how trading activity is fluctuating over time.

By combining these different types of features, the dataset now contains a richer representation of the Bitcoin market's state at each timestamp, enabling the models to learn more complex patterns and relationships for improved predictive performance.

## Post-Feature Engineering - Distribution Analysis

### Distribution of Engineered Features

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import scipy.stats as stats

def plot_distribution(df_pd, feature_name):
    """
    Generates a triple plot (histogram/KDE, box plot, Q-Q plot) for a given feature.

    Args:
        df_pd (pd.DataFrame): Pandas DataFrame containing the data.
        feature_name (str): Name of the feature column to plot.
    """
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    fig.suptitle(f'Distribution Analysis for {feature_name}', fontsize=16)

    # Subplot 1: Histogram with KDE
    sns.histplot(df_pd[feature_name], kde=True, ax=axes[0])
    axes[0].set_title('Histogram with KDE')
    axes[0].set_xlabel(feature_name)
    axes[0].set_ylabel('Frequency')

    # Subplot 2: Box Plot
    sns.boxplot(x=df_pd[feature_name], ax=axes[1])
    axes[1].set_title('Box Plot')
    axes[1].set_xlabel(feature_name)

    # Subplot 3: Q-Q Plot
    stats.probplot(df_pd[feature_name], dist="norm", plot=axes[2])
    axes[2].set_title('Q-Q Plot')

    plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent title overlap
    plt.show()

# Convert the necessary features from the final Spark DataFrame to Pandas for plotting
# Select a sample if the DataFrame is too large for direct conversion
# For demonstration, let's select a subset of the engineered features
features_to_plot = ["Close_MA30", "Volume_Lag1"]

# It's recommended to sample a large Spark DataFrame before converting to Pandas
# For example, if df_final is very large, use:
# df_final_sampled_pd = df_final.select(features_to_plot).sample(False, 0.1, seed=42).toPandas()
# For this task, assuming df_final_engineered is manageable in memory:
df_final_pd = df_engineered.select(features_to_plot).toPandas()


# Call the plotting function for the selected features
for feature in features_to_plot:
    plot_distribution(df_final_pd, feature)
```

## Data Partitioning

Chronologically spliting the engineered DataFrame into training and testing sets.

```python
from pyspark.sql.functions import to_date
from pyspark.ml.feature import VectorAssembler

split_date = "2020-01-01"

train_df = df_final.filter(to_date(col("Timestamp")) < split_date)
test_df = df_final.filter(to_date(col("Timestamp")) >= split_date)

print(f"Training set size: {train_df.count()} rows")
print(f"Testing set size: {test_df.count()} rows")

# Prepare features for ML models
# Exclude Timestamp and target columns from the feature list
feature_columns_ml = [col for col in feature_columns if col != "Timestamp"]

assembler = VectorAssembler(inputCols=feature_columns_ml, outputCol="features")

train_df_assembled = assembler.transform(train_df)
test_df_assembled = assembler.transform(test_df)

# Cache the assembled DataFrames as they will be used multiple times
train_df_assembled.cache()
test_df_assembled.cache()

print("\nSchema of training set with features vector:")
train_df_assembled.printSchema()

print("\nFirst 5 rows of training set with features vector:")
train_df_assembled.show(5)
```

**Reasoning**:
For time-series data, a chronological split is crucial because it preserves the temporal order of observations. Training on past data and testing on future data simulates a real-world scenario where models are used to make predictions on unseen, future information. Using a random split would introduce data leakage, where information from the future could be included in the training set, leading to overly optimistic performance estimates and poor generalization to true future data.

## Regression Task

Train Linear Regression and GBT Regression models to predict the next minute's closing price (`Close_t+1`).

```python
from pyspark.ml.regression import LinearRegression, GBTRegressor
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import RegressionEvaluator

# Define target variable for regression
regression_target = "Close_t+1"

# --- Linear Regression Model ---
lr = LinearRegression(featuresCol="features", labelCol=regression_target)
lr_model = lr.fit(train_df_assembled)

print("\nLinear Regression Model trained.")

# Evaluate Linear Regression Model
lr_predictions = lr_model.transform(test_df_assembled)
regression_evaluator = RegressionEvaluator(labelCol=regression_target, predictionCol="prediction", metricName="rmse")
lr_rmse = regression_evaluator.evaluate(lr_predictions)
print(f"Linear Regression RMSE on test set: {lr_rmse}")

# --- GBT Regression Model ---
# Using a subset of data for GBTRegressor training due to potential memory constraints with large datasets
# For a full dataset, consider using distributed strategies or reducing the number of trees/depth
gbt_r = GBTRegressor(featuresCol="features", labelCol=regression_target, maxIter=10) # Reduced iterations for demonstration
gbt_r_model = gbt_r.fit(train_df_assembled.sample(False, 0.1, seed=42)) # Sampled data for faster training

print("\nGBT Regression Model trained (on sampled data).")

# Evaluate GBT Regression Model
gbt_r_predictions = gbt_r_model.transform(test_df_assembled)
gbt_r_rmse = regression_evaluator.evaluate(gbt_r_predictions)
print(f"GBT Regression RMSE on test set: {gbt_r_rmse}")
```

## Classification Task

Train Logistic Regression and GBT Classification models to predict the direction of price change (`Direction_t+1`).

```python
from pyspark.ml.classification import LogisticRegression, GBTClassifier
from pyspark.ml.evaluation import BinaryClassificationEvaluator

# Define target variable for classification
classification_target = "Direction_t+1"

# --- Logistic Regression Model ---
lr_c = LogisticRegression(featuresCol="features", labelCol=classification_target)
lr_c_model = lr_c.fit(train_df_assembled)

print("\nLogistic Regression Model trained.")

# Evaluate Logistic Regression Model
lr_c_predictions = lr_c_model.transform(test_df_assembled)
binary_evaluator = BinaryClassificationEvaluator(labelCol=classification_target, rawPredictionCol="rawPrediction", metricName="areaUnderROC")
lr_c_auc = binary_evaluator.evaluate(lr_c_predictions)
print(f"Logistic Regression Area under ROC on test set: {lr_c_auc}")

# --- GBT Classification Model ---
# Using a subset of data for GBTClassifier training due to potential memory constraints with large datasets
gbt_c = GBTClassifier(featuresCol="features", labelCol=classification_target, maxIter=10) # Reduced iterations for demonstration
gbt_c_model = gbt_c.fit(train_df_assembled.sample(False, 0.1, seed=42)) # Sampled data for faster training

print("\nGBT Classification Model trained (on sampled data).")

# Evaluate GBT Classification Model
gbt_c_predictions = gbt_c_model.transform(test_df_assembled)
gbt_c_auc = binary_evaluator.evaluate(gbt_c_predictions)
print(f"GBT Classification Area under ROC on test set: {gbt_c_auc}")
```

## Model Evaluation

Evaluating the performance of our trained models on the test set.

```python
from pyspark.ml.evaluation import RegressionEvaluator, MulticlassClassificationEvaluator

# Regression Model Evaluation (RMSE and MAE)
regression_evaluator_rmse = RegressionEvaluator(labelCol=regression_target, predictionCol="prediction", metricName="rmse")
regression_evaluator_mae = RegressionEvaluator(labelCol=regression_target, predictionCol="prediction", metricName="mae")

lr_rmse = regression_evaluator_rmse.evaluate(lr_predictions)
lr_mae = regression_evaluator_mae.evaluate(lr_predictions)

gbt_r_rmse = regression_evaluator_rmse.evaluate(gbt_r_predictions)
gbt_r_mae = regression_evaluator_mae.evaluate(gbt_r_predictions)

print("--- Regression Model Evaluation ---")
print(f"Linear Regression - RMSE: {lr_rmse:.4f}, MAE: {lr_mae:.4f}")
print(f"GBT Regression (sampled) - RMSE: {gbt_r_rmse:.4f}, MAE: {gbt_r_mae:.4f}")

# Classification Model Evaluation (Accuracy and F1-Score)
# Note: MulticlassClassificationEvaluator requires prediction and label columns to be double type
classification_evaluator_accuracy = MulticlassClassificationEvaluator(labelCol=classification_target, predictionCol="prediction", metricName="accuracy")
classification_evaluator_f1 = MulticlassClassificationEvaluator(labelCol=classification_target, predictionCol="prediction", metricName="f1")

# Cast label and prediction columns to double for MulticlassClassificationEvaluator
lr_c_predictions_double = lr_c_predictions.withColumn(classification_target, col(classification_target).cast("double")).withColumn("prediction", col("prediction").cast("double"))
gbt_c_predictions_double = gbt_c_predictions.withColumn(classification_target, col(classification_target).cast("double")).withColumn("prediction", col("prediction").cast("double"))


lr_c_accuracy = classification_evaluator_accuracy.evaluate(lr_c_predictions_double)
lr_c_f1 = classification_evaluator_f1.evaluate(lr_c_predictions_double)

gbt_c_accuracy = classification_evaluator_accuracy.evaluate(gbt_c_predictions_double)
gbt_c_f1 = classification_evaluator_f1.evaluate(gbt_c_predictions_double)


print("\n--- Classification Model Evaluation ---")
print(f"Logistic Regression - Accuracy: {lr_c_accuracy:.4f}, F1-Score: {lr_c_f1:.4f}")
print(f"GBT Classification (sampled) - Accuracy: {gbt_c_accuracy:.4f}, F1-Score: {gbt_c_f1:.4f}")
```
